```python
def get_pandas_dataframe_from_uncooperative_hdf5(filename, key='first_key'):
    f = h5py.File(filename,'r')
    all_keys = list(f.keys())
    
    if key == 'first_key':
        print('Dataset contains these keys: ')
        print(all_keys)
        key = all_keys[0]
        print('Using only the first key: ', key)
        
    data = f[key][()]
    dic = {}
    for column_label in data.dtype.fields.keys():
        dic.setdefault(column_label, data[column_label])
    df = pandas.DataFrame(dic)
    return df
```

```python
def get_wind_speed_and_turbulence_for_time(wind_df, start_time,wind_in,Volt_in, delay=6, time_range=30):
    # extract part of data frame corresponding to that time range
    # then do whatever averaging, std, etc.
    
    start_time_delayed = start_time + delay
    end_time = start_time + time_range

    q = "t_secs > " + str(start_time_delayed) + " and " + "t_secs < " + str(end_time)
    df_tmp = df.query(q)
    
    #NS = df_tmp['northsouth']
    #WE = df_tmp['westeast']
    #UD = df_tmp['updown']
    
    #interpulate
    z=np.interp(df_tmp.data_0,Volt_in,wind_in)
    # now calculate means, std
    wind_3d = z
    mean_wind_3d = np.mean(wind_3d)
    
    std_wind_3d = np.mean(np.std(wind_3d))
    turbulent_intensity = std_wind_3d / mean_wind_3d
    
    return mean_wind_3d, turbulent_intensity
```

```python
def get_position_and_wind_characteristics_for_row(row, mf, df):
    position_data = mf.iloc[row]
    x_pos = position_data.data_0
    y_pos = position_data.data_1
    start_time = position_data.t_secs
    mean_wind, turb_intensity = get_wind_speed_and_turbulence_for_time(df, start_time,wind,VA)

    return x_pos, y_pos, mean_wind, turb_intensity
```

```python
X = []
Y = []
MW = []
TI = []

for row in range(mf.shape[0]):
    x_pos, y_pos, mean_wind, turb_intensity = get_position_and_wind_characteristics_for_row(row, mf, df)
    X.append(x_pos)
    Y.append(y_pos)
    MW.append(mean_wind)
    TI.append(turb_intensity)
```

```python
X = np.array(X)
Y = np.array(Y)
MW = np.array(MW)
TI=np.array(TI)
```

```python
# calibrate steps

cX = (X - X.min())
cX /= cX.max()
cX *= 15*2.54 # cm

cY = (Y - Y.min())
cY /= cY.max()
cY *= 13*2.54 # cm 
```

```python
from math import pow  
from math import sqrt  
import numpy as np  
import matplotlib.pyplot as plt  

def pointValue(x,y,power,smoothing,xv,yv,values):  
    nominator=0  
    denominator=0  
    for i in range(0,len(values)):  
        dist = sqrt((x-xv[i])*(x-xv[i])+(y-yv[i])*(y-yv[i])+smoothing*smoothing);  
        #If the point is really close to one of the data points, return the data point value to avoid singularities  
        if(dist<0.0000000001):  
            return values[i]  
        nominator=nominator+(values[i]/pow(dist,power))  
        denominator=denominator+(1/pow(dist,power))  
    #Return NODATA if the denominator is zero  
    if denominator > 0:  
        value = nominator/denominator  
    else:  
        value = -9999  
    return value  

def invDist(xv,yv,values,xsize=100,ysize=100,power=2,smoothing=0):  
    valuesGrid = np.zeros((ysize,xsize))  
    for x in range(0,xsize):  
        for y in range(0,ysize):  
            valuesGrid[y][x] = pointValue(x,y,power,smoothing,xv,yv,values)  
    return valuesGrid  

```

```python
ti = np.linspace(np.min((cX.min(), cY.min())), np.max((cX.max(), cY.max())), 100)  
```

```python
power=5
smoothing=5  

#Creating the output grid (100x100, in the example)  
iX, iY = np.meshgrid(ti, ti)  

#Creating the interpolation function and populating the output matrix value  
iMW = invDist(cX,cY,MW,100,100,power,smoothing) 
iTI= invDist(cX,cY,TI,100,100,power,smoothing)
```
